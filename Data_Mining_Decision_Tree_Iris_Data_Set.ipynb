## 4. Calculate the Euclidean distance between all records and the mean of each class

I am adding new columns to a copy of the original dataframe. For eg, data_copy['mean_setosa_dist'] is a new column in dataframe 'data_copy' that has the Euclidian distance between  each record and the mean sepal length(sl), sepal width(sw), petal length(pl), and petal width(pw) of Setosa (there were found in q3 and saved in variables like set_sl_mean(setosa sepal length mean)). 

## Math notation would look like: 
   <b> √ (sl - set_sl_mean)^2 + (sw - set_sw_mean)^2 + (pl - set_pl_mean)^2 + (pw - set_pw_mean)^2 </b>

```python
data_copy = data.copy()
```

### a. Setosa

```python
data_copy['mean_setosa_dist'] = np.sqrt(np.square(data_copy['sepal length'] - set_sl_mean) + np.square(data_copy['sepal width'] - set_sw_mean)+ np.square(data_copy['petal length'] - set_pl_mean) + np.square(data_copy['petal width'] - set_pw_mean))
```

```python
data_copy.head()
```

### b. Versicolor

```python
data_copy['mean_versicolor_dist'] = np.sqrt(np.square(data_copy['sepal length'] - versi_sl_mean) + np.square(data_copy['sepal width'] - versi_sw_mean)+ np.square(data_copy['petal length'] - versi_pl_mean) + np.square(data_copy['petal width'] - versi_pw_mean))
```

```python
data_copy.head()
```

### c. Virginica

```python
data_copy['mean_virginica_dist'] = np.sqrt(np.square(data_copy['sepal length'] - virg_sl_mean) + np.square(data_copy['sepal width'] - virg_sw_mean)+ np.square(data_copy['petal length'] - virg_pl_mean) + np.square(data_copy['petal width'] - virg_pw_mean))
```

### Using matplotlib

With matplotlib we can use pyplot to create the visualisations. For this we must first create the required combinations of attributes and we use the `itertools.combinations(iterable, r)` function from the library `itertools`. 

The `scatter()` does not allow for different marker types based on the class attribute. To achieve different markers we need to separate the series based on class and plot 3 times on the same subplot. 

Links to itertools: https://docs.python.org/3/library/itertools.html#itertools.combinations


```python
names = list(data.columns)
names.remove("class")
comb = list(itertools.combinations(names, 2))

fig, axs = plt.subplots(2, 3, figsize=(12,8))


for i in range(0, 2):
    for j in range(0,3):
        axs[i,j].scatter(data[data["class"] == "Iris-versicolor"][comb[(3*i+j)][0]],
                     data[data["class"] == "Iris-versicolor"][comb[(3*i+j)][1]], 
                c = "red",
               marker = "x")
        axs[i,j].scatter(data[data["class"] == "Iris-setosa"][comb[(3*i+j)][0]],
                     data[data["class"] == "Iris-setosa"][comb[(3*i+j)][1]], 
                c = "blue",
               marker = "o")

        axs[i,j].scatter(data[data["class"] == "Iris-virginica"][comb[(3*i+j)][0]],
                     data[data["class"] == "Iris-virginica"][comb[(3*i+j)][1]], 
                c = "green",
               marker = "+")
        axs[i,j].set_xlabel(comb[(3*i+j)][0])
        axs[i,j].set_ylabel(comb[(3*i+j)][1])
        

plt.legend()
plt.tight_layout()
```

## 3D Plots

3D Plots require `Axes 3D` in addition to matplotlib

```python
from mpl_toolkits.mplot3d import Axes3D

comb = list(itertools.combinations(names, 3))

fig, axs = plt.subplots(2, 2, subplot_kw={'projection':'3d'}, figsize=(12,8))


for i in range(0, 2):
    for j in range(0,2):
        axs[i,j].scatter(data[data["class"] == "Iris-versicolor"][comb[(2*i+j)][0]],
                    data[data["class"] == "Iris-versicolor"][comb[(2*i+j)][1]],
                    data[data["class"] == "Iris-versicolor"][comb[(2*i+j)][2]],    
                    c = "red", marker = "x", label="Iris-versicolor")
        axs[i,j].scatter(data[data["class"] == "Iris-setosa"][comb[(2*i+j)][0]],
                    data[data["class"] == "Iris-setosa"][comb[(2*i+j)][1]],
                    data[data["class"] == "Iris-setosa"][comb[(2*i+j)][2]],
                    c = "blue", marker = "o", label="Iris-setosa")

        axs[i,j].scatter(data[data["class"] == "Iris-virginica"][comb[(2*i+j)][0]],
                    data[data["class"] == "Iris-virginica"][comb[(2*i+j)][1]],
                    data[data["class"] == "Iris-virginica"][comb[(2*i+j)][2]],
                    c = "green", marker = "+", label="Iris-virginica")
        
        axs[i,j].set_xlabel(comb[(2*i+j)][0])
        axs[i,j].set_ylabel(comb[(2*i+j)][1])
        axs[i,j].set_zlabel(comb[(2*i+j)][2])
        
plt.legend()
plt.tight_layout()
```
